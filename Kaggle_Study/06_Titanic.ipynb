```python
import kagglehub

# Download latest version
path = kagglehub.competition_download('titanic')

print("Path to competition files:", path)
```

```python
import os

for dirname, _, filenames in os.walk(path):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
import pandas as pd

train_df = pd.read_csv(os.path.join(path, "train.csv"))
test_df = pd.read_csv(os.path.join(path, "test.csv"))

train_df.head()
```

```python
predictions = []
for sex in test_df['Sex']:
    if sex == 'female':
        predictions.append(1)
    else:
        predictions.append(0)

print(len(predictions))
```

```python
# 1. 캐글이 요구한 컬럼명으로 새로운 딕셔너리 만들기
submission_data = {
    'PassengerId': test_df['PassengerId'], # 시험지에 있던 승객 번호 그대로
    'Survived': predictions # 방금 우리가 예측한 0 또는 1 데이터
}

# 2. 판다스 데이터프레임(표)으로 변환
submission_df = pd.DataFrame(submission_data)

# 3. *이 코드가 핵심!* 구글 코랩에 'my_first_submission.csv'라는 파일로 내보내기
# index=False를 안 해주면 쓸데없는 앞 숫자가 같이 저장되어 캐글에서 에러가 납니다.
submission_df.to_csv('my_first_submission.csv', index=False)
```
