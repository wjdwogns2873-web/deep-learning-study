```python
import torch
import torch.nn as nn
import torchvision.models as models
device = torch.device('mps' if torch.backends.mps.is_available() else 'cpu')

model = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)

num_features = model.fc.in_features
model.fc = nn.Linear(num_features, 6)

model = model.to(device)

model.load_state_dict(torch.load('fold4_100.pth', map_location=device))
model.eval()
print('complete4')
```

```python
import albumentations as A
from albumentations.pytorch import ToTensorV2
print("ff")
test_transform = A.Compose([
    A.Resize(128, 128), 
    A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)), 
    ToTensorV2()
])
```

```python
import os
wrong_dir = "images/kaggle_data/dataset/fold4_acc_100_wrong_image"
os.makedirs(wrong_dir, exist_ok=True)
```

```python
import os
from PIL import Image
import numpy as np
import shutil
count, correct_count, wrong_count = 0, 0, 0

classes = []

fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    classes.append(f"fresh_{fruit}")
    classes.append(f"rotten_{fruit}")

for fruit in fruits:
    path = f"images/kaggle_data/dataset/{fruit}"
    categories = [f"fresh_{fruit}", f"rotten_{fruit}"]
    
    for category in categories:
        folder_path = os.path.join(path, category)
        for image in os.listdir(folder_path):
            if image.endswith(('jpg', 'jpeg', 'png')):
                count += 1
                if count % 1000 == 0:
                    print(f"count: {count}")
                full_path = os.path.join(folder_path, image)
    
                pil_image = Image.open(full_path).convert('RGB')
                np_image = np.array(pil_image)
    
                aug = test_transform(image=np_image)
                input_tensor = aug['image'].unsqueeze(0)
    
                with torch.no_grad():
                    outputs = model(input_tensor.to(device))
                    pred = outputs.argmax(1).item()
    
                pred_text = classes[pred]
    
                if pred_text == category:
                    correct_count += 1
                elif pred_text != category:
                    wrong_count += 1
                    shutil.copy(full_path, os.path.join(wrong_dir, image))

print(f"Total: {count}, Correct: {correct_count}, Wrong: {wrong_count}")
```
