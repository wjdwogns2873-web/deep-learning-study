```python
# nn.Module 상속 클래스 정의
# __init__(): 모델에서 사용될 모듈과 활성화 함수 등을 정의
# forward(): 모델에서 실행되어야 하는 연산을 정의
class Model(nn.Module):
    def __init__(self, inputs):
        super(Model, self).__init__()
        self.layer = nn.Linear(inputs, 1)
        self.activation = nn.Sigmoid()

    def forward(self, x):
        x = self.layer(x)
        x = self.activation(x)
        return x
```

```python
# 사용 예
# 1. 모델 생성 (입력 특징 개수가 784인 모델입니다잉!)
my_model = Model(inputs=784).to(device)
# 2. 가상의 입력 데이터 (Batch size = 4)
example_input = torch.randn(4, 784).to(device)
# 3. 예측 수행 (forward 실행)
prediction = my_model(example_input)
print(prediction.shape)
```

```python
# nn.Sequential()을 이용한 신경망 정의
# nn.Sequential 객체로 그 안에 각 모듈을 순차적으로 실행
# __init__()에서 사용할 네트워크 모델들을 nn.Sequential로 정의 가능
# forward()에서 실행되어야 할 계산을 가독성 높게 작성 가능
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=64, kernel_size=5),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2)
        )

        self.layer2 = nn.Sequential(
            nn.Conv2d(in_channels=64, out_channels=30, kernel_size=5),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2)
        )

        self.layer3 = nn.Sequential(
            nn.Linear(in_features=30*5*5, out_features=10, bias=True),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        x = self.layer1(x)
        x = self.layer2(x)
        x = x.view(x.shape[0], -1)
        x = self.layer3(x)
        return x
```

```python
# 클래스 초기화 시 task와 함께 num_class를 명시해야 합니다.
metrics = torchmetrics.Accuracy(task="multiclass", num_classes=5)

n_batches = 10
for i in range(n_batches):
    preds = torch.randn(10, 5).softmax(dim=1)
    target = torch.randint(5, (10, ))

    # 현재 배치에 대한 정확도 계산
    acc = metrics(preds, target)
    print(f"Batch {i} Accuracy: {acc:.4f}")

# 전체 배치가 끝난 후 '전체 평균 정확도' 계산
total_acc= metrics.compute()
print(f"\nTotal Average Accuracy: {total_acc:.4f}")

# 계산이 끝난 후 다음 에폭에서 사용하려면 리셋 (선택사항)
metrics.reset()
```

```python
import matplotlib.pyplot as plt
# 데이터 생성
X = torch.randn(200, 1) * 10
y = X + 3 * torch.randn(200, 1)
plt.scatter(X.numpy(), y.numpy())
plt.ylabel('y')
plt.xlabel('X')
plt.grid()
plt.show()
```

```python
# 모델 정의 및 파라미터
import torch.nn as nn
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super(LinearRegressionModel, self).__init__()
        self.linear = nn.Linear(1, 1)
    def forward(self, x):
        # y = ax + b 연산이 여기서 일어남.
        pred = self.linear(x)
        return pred
```

```python
model = LinearRegressionModel()
print(model)
print(list(model.parameters()))
```

```python
import numpy as np
w, b = model.parameters()
# 현재 모델은 nn.Linear(1, 1)이므로 weight의 shape은 [1, 1]입니다. [[0.1234]]
w1, b1 = w[0][0].item(), b[0].item()
x1 = np.array([-30, 30])
y1 = w1 * x1 + b1
plt.plot(x1, y1, 'r')
plt.scatter(X, y)
plt.grid()
plt.show()
```

```python
# 손실 함수 및 옵티마이저
import torch.optim as optim

criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.001)
```

```python
# 모델 학습
epochs = 100
losses = []

for epoch in range(epochs):
    optimizer.zero_grad()

    y_pred = model(X)
    loss = criterion(y_pred, y)
    losses.append(loss.item())
    loss.backward()

    optimizer.step()
```

```python
plt.plot(range(epochs), losses)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()
```

```python
import torch.nn.functional as F
class NeuralNet(nn.Module):
    def __init__(self):
        super(NeuralNet, self).__init__()

        self.conv1 = nn.Conv2d(1, 6, 3)
        self.conv2 = nn.Conv2d(6, 16, 3)
        self.fc1 = nn.Linear(16 * 5 * 5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)
    def forward(self, x):
        x = F.max_pool2d(F.relu(self.conv1(x)), (2, 2))
        x = F.max_pool2d(F.relu(self.conv2(x)), 2)
        # num_flat_features함수는 배치 차원을 제외한 나머지 차원들을 하나로 곱해주는 함수.
        x = x.view(-1, self.num_flat_features(x))
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
    def num_flat_features(self, x):
        size = x.size()[1:]
        num_features = 1
        for s in size:
            num_features *= s

        return num_features

net = NeuralNet()
print(net)
```

```python
# 모델 테스트
def imshow(image):
    # 역정규화 과정
    # 처음에 Normalize((0.5, ), (0.5, ))를 해서 -1, 1 사이로 만든 것을 되돌림.
    # 이미지를 화면에 뿌릴 때는 0~1의 값으로 되돌려야 색상이 제대로 보입니다.
    image = image / 2 + 0.5
    npimg = image.numpy()

    fig = plt.figure(figsize=(16, 8))
    # PyTorch: (채널, 가로, 세로) 순서
    # Matplotlib: (가로, 세로, 채널) 순서
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()
```
